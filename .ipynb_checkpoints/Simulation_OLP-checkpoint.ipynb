## SIMULATION_OLP

```python
import pandas as pd
import numpy as np
import math
import package_DBR

import matplotlib.pyplot as plt
from package_DBR import myRound, SelectPath_RT, Delay_RT, FO_RT, FOPDT, SOPDT, FOPDT_cost, SOPDT_cost, Process, Bode
```

### Load Data

```python
nameFile = 'Cleaned_data_Open_loop_experiment_on_MV_2024-03-18-09h26.txt'

titleName = nameFile.split('.')[0]    
data = pd.read_csv('Data/' + nameFile)

if 'MV' in nameFile:
    ExpVariable = 'MV'
    tm = data['tm'].values
    MVm = data['MVm'].values
    PVm = data['PVm'].values    
else:    
    ExpVariable = 'DV'
    tm = data['tm'].values
    DVm = data['DVm'].values 
    PVm = data['PVm'].values
     
print(ExpVariable)    
```

### Model Approximations

```python
a = 0.12   # [째C/%]
t1 = 77    # 28% of PV_inf [sec]
t2 = 107   # 40% of PV_inf [sec]
T_u = 16   # Approximated Delay [sec]
T_g = 244  # 
Kp = 0.47  # % of PV value [째C/%]

PV_28 = 0.4*Kp # % of PV value [째C/%]
PV_40 = 0.4*Kp # % of PV value [째C/%]

print(PV_28)
print(PV_40)

t = []
MV = []

TSim = 720 # Simulation Time [s]

MVPath = {0:0, 1:1, TSim:1}
Ts = 1     # Sampling Period [s]

N = int(TSim/Ts) + 1

for i in range(0,N):
    t.append(i*Ts)
    SelectPath_RT(MVPath,t,MV)
```

### Response for FOPDT (obtained with Broida Model 1)

```python
T_B1 = T_g
theta_B1 = T_u
```

### Response for FOPDT (obtained with Broida Model 2)

```python
T_B2 = 5.5*(t2 - t1) 
theta_B2 = (2.8*t1) - (1.8*t2)

print(T_B2, theta_B2)
```

### Response for SOPDT (obtained with van der Grinten)

```python
#cstes for VDG model
e = math.e
#T1VDG = T_g * (3 * a * e - 1)/(1 + a * e)
T1VDG = 0
T2VDG = T_g * (1 - a * e)/(1 + a * e)
thetaVDG = T_u - (T1VDG * T2VDG)/(T1VDG + 3 * T2VDG)

print(T1VDG, T2VDG, thetaVDG)
```

### Response for n-th Order model with Identical Poles (obtained with Strejc)

```python
a_n_test = T_u / T_g # = 0.0655 => a_n = 0 =< 0.0655 < a_n+1 = 1  => n = 1
print(a_n_test)

n = 1
a_n = 0
b_n = 1

T_T = T_g / b_n
T_u_th_T = a_n * T_g
theta_T = T_u - T_u_th_T

base_T = Ts + 1

print(T_T, theta_T)
```

### Step Responses

```python
FOPDT_P_B1 = FOPDT(MVm, Kp, T_B1, theta_B1, Ts)
FOPDT_P_B2 = FOPDT(MVm, Kp, T_B2, theta_B2, Ts)
PVSOPDT_VDG = SOPDT(MVm,Kp,0,T2VDG,thetaVDG,Ts)
FOPDT_P_T = FOPDT(MVm, Kp, T_T, theta_T, Ts)


plt.figure(figsize = (15,9))

plt.subplot(2,1,2)
plt.step(tm,PVm,'green',label='Real Process',where='post')
plt.step(tm,FOPDT_P_B1,'cyan',label='FOPDT Broida 1',where='post')
plt.step(tm,FOPDT_P_B2,'red',label='FOPDT Broida 2',where='post')
plt.step(tm,PVSOPDT_VDG,'lime',label='SOPDT van der Grinten', where='post')
plt.step(tm,FOPDT_P_T,'blue',label='FOPDT Strejc', ls = ':',where='post')
plt.ylabel('Value of PV')
plt.xlabel('Time [s]')
plt.legend(loc='best')
plt.xlim([0, TSim])    
```

### Bode Diagrams

```python
P_B1 = Process({}) # Default process = unit gain
P_B2 = Process({}) # Default process = unit gain
P_VDG = Process({}) # Default process = unit gain
P_S = Process({}) # Default process = unit gain

# Parameters for input-output dynamics P(s)
P_B1.parameters['Kp'] = Kp
P_B1.parameters['Tlag1'] = T_B1
P_B1.parameters['Tlag2'] = 0
P_B1.parameters['theta'] = theta_B1

# Parameters for input-output dynamics P(s)
P_B2.parameters['Kp'] = Kp
P_B2.parameters['Tlag1'] = T_B2
P_B2.parameters['Tlag2'] = 0
P_B2.parameters['theta'] = theta_B2

# Parameters for input-output dynamics P(s)
P_VDG.parameters['Kp'] = Kp
P_VDG.parameters['Tlag1'] = 0
P_VDG.parameters['Tlag2'] = T2VDG
P_VDG.parameters['theta'] = thetaVDG

# Parameters for input-output dynamics P(s)
P_S.parameters['Kp'] = Kp
P_S.parameters['Tlag1'] = T_T
P_S.parameters['Tlag2'] = 0
P_S.parameters['theta'] = theta_T


omega = np.logspace(-4, 1, 10000)

P1 = Bode(P_B1,omega,False)
P2 = Bode(P_B2,omega,False)
P3 = Bode(P_VDG,omega,False)
P4 = Bode(P_S,omega,False)


plt.subplot(2,1,1)
gain_min = np.min(20*np.log10(np.abs(P1)/5))
gain_max = np.max(20*np.log10(np.abs(P1)*5))
plt.semilogx(omega,20*np.log10(np.abs(P1)),'cyan',ls = '-', linewidth = 2,label='Broida model 1')
plt.semilogx(omega,20*np.log10(np.abs(P2)),'red',linewidth=2,label='Broida model 2')
plt.semilogx(omega,20*np.log10(np.abs(P3)),'lime',linewidth=2,label='van der Grinten')
plt.semilogx(omega,20*np.log10(np.abs(P4)),'blue',ls = ':',linewidth=2,label='Strejc')
plt.xlim([np.min(omega), np.max(omega)])
plt.ylim([gain_min, gain_max])
plt.ylabel('Amplitude [db]')
plt.legend(loc='best')

plt.subplot(2,1,2)
ph_min = np.min((180/np.pi)*np.unwrap(np.angle(P1))) - 10
ph_max = np.max((180/np.pi)*np.unwrap(np.angle(P1))) + 10
plt.semilogx(omega, (180/np.pi)*np.unwrap(np.angle(P1)),'cyan',linewidth=2)
plt.semilogx(omega, (180/np.pi)*np.unwrap(np.angle(P2)),'red',linewidth=2)
plt.semilogx(omega, (180/np.pi)*np.unwrap(np.angle(P3)),'lime',linewidth=2)
plt.semilogx(omega, (180/np.pi)*np.unwrap(np.angle(P4)),'blue',ls = ':',linewidth=2)
plt.xlim([np.min(omega), np.max(omega)])
plt.ylim([np.max([ph_min, -200]), ph_max])
plt.ylabel('Phase [째]')
```
